```r
data <- read.csv("college_vs_w.csv", header=TRUE)
```

```r
colnames(data)
```

```r
#college FT% vs WNBA 3p%
#percentiles and raw

plot(data$P_CFT., data$P_W3p., pch = 16, col = "black")
cor(data$P_CFT., data$P_W3p.)

plot(data$CFT., data$W3p., pch = 16, col = "black", main="Player College FT% vs WNBA 3p%", xlab="College Career FT%", ylab="WNBA 2023 3p%")
cor(data$CFT., data$W3p.)
```

```r
#college 3p% vs WNBA 3p%
#percentiles and raw

plot(data$P_C3p., data$P_W3p., pch = 16, col = "black")
cor(data$P_C3p., data$P_W3p.)

plot(data$C3p., data$W3p., pch = 16, col = "black",main="Player College 3p% vs WNBA 3p%", xlab="College Career 3p%", ylab="WNBA 2023 3p%")
cor(data$C3p., data$W3p.)
fit <- lm(data$W3p. ~ data$C3p.)
abline(fit, col = "red")
```
